### Loading data

```python
import datacube

dc = datacube.Datacube(app="my_analysis")
ds = dc.load(product="diwata_2_smi",
             x=(119.6246, 121.0693),
             y =(15.9613, 18.7867),
             time=("2019-01-01", "2019-12-31"),
             output_crs='EPSG:32651',
             resolution=(-122, 122)
            )


ds
```

```python
ds1 = ds.where(ds > 0)
```

```python
import numpy as np
import matplotlib.pyplot as plt

rgb = np.dstack((ds1.band3.isel(time=3).values,
                 ds1.band2.isel(time=3).values,
                 ds1.band1.isel(time=3).values))

plt.figure(figsize = (20,10))
plt.imshow(rgb)
```

```python
ndvi = (ds1.band6 - ds1.band3) / (ds1.band6 + ds1.band3)
plt.figure(figsize = (20,10))
plt.imshow(ndvi.isel(time=3).values, cmap="RdYlBu")
plt.clim(-0.054,0.84)
plt.colorbar()
plt.show()
```
